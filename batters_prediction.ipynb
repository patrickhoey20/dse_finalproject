```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("batters.csv")
```

```python
df = df.replace(False, 0)
df = df.replace(True, 1)
df.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
batters_data = df[df.active_player ==1]
batters_data.head()
```
